```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../../../Documents/01_ACTIVE_PROJECTS\
/Predicteds_and_Cuts/Line_asignment/SME/apps_finalv1.csv').iloc[:, 1:]
```

```python
df[(df.state == 'A')].groupby('new_risk_tier').agg(
        {'notional':'sum'}).style.format('{:,.0f}')
```

```python
#dictionaries to format
font_style = {
    'fontsize':'xx-large',
    'fontfamily':'sans-serif',
    'style':'oblique',
    'weight':'normal'}

ticks_style = {
    'fontsize':'x-large',
    'fontfamily':'sans-serif',
    'style':'oblique',
    'weight':'light'}
```

```python
fig, ax = plt.subplots(figsize = (20, 10))

df[(df.state == 'A')].groupby('new_risk_tier').agg(
        {'notional':'sum'}).loc[
    ['AAA', 'A', 'A-', 'B', 'C']
].plot(
    kind = 'bar', 
    ax = ax,
    legend = True,
    zorder = 3,
    color='tab:blue'
)

ax.legend(["Legend label"],
          fontsize='x-large')

ax.set_title(label = 'Notional by Risk Tier', fontdict = font_style)
ax.set_xlabel('Risk tiers', fontdict = font_style)
ax.set_ylabel('Notinoal', fontdict = font_style)

ax.set_xticklabels(
    ['Mejor', 'Bueno', 'Apenas', 'Malo', 'Horroso'],
    fontdict=ticks_style,
    rotation = 45)


ax.set_yticks(np.arange(0, 160_000_000, 20_000_000))
ax.set_yticklabels(
    labels = ['${:,.0f} MM'.format(i) for i in np.arange(0, 160, 20)],
    fontdict=ticks_style)

ax.grid(axis = 'y',
        alpha = 1,
        linestyle='-.',
        zorder=0)
```
